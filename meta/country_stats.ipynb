```python
import pandas as pd
import numpy as np
from datetime import datetime,timedelta
```

```python
countries = pd.read_csv('https://raw.githubusercontent.com/kazuhirokida/classification/master/JHU_Classification.csv',na_filter=False)
```

```python
jhu = pd.read_csv('https://github.com/kazuhirokida/covid-19/raw/master/countries.csv',na_filter=False)
```

```python
iso = pd.read_csv('https://raw.githubusercontent.com/kazuhirokida/classification/master/ISO.csv',na_filter=False)
```

```python
countries['Japanese'] = countries.ISO2.map(jhu.set_index('Code').Japanese.to_dict())
countries['latlong'] = countries.ISO2.map(jhu.set_index('Code').latlong.to_dict())
countries['Numeric'] = countries.ISO2.map(iso.set_index('Alpha-2 code').Numeric.to_dict())
```

```python
countries = countries.append(pd.Series({'ISO2':'VU','ISO3':'VUT','Country':'Vanuatu','Region':'Oceania',
                      'Classification':'Developing','Japanese':'バヌアツ','latlong':'-17.734,168.323','Numeric':548}),ignore_index=True)
```

```python
countries = countries.append(pd.Series({'ISO2':'WS','ISO3':'WSM','Country':'Samoa','Region':'Oceania',
                      'Classification':'Developing','Japanese':'サモア','latlong':'-13.833,-171.75','Numeric':882}),ignore_index=True)
```

```python
countries
```

```python
un = pd.read_csv('https://population.un.org/wpp/Download/Files/1_Indicators%20(Standard)/CSV_FILES/WPP2019_TotalPopulationBySex.csv')
un = un[(un.Time==2020)&(un.Variant=='Medium')]
```

```python
countries['Region_j'] = countries.Region.map(
    {'Africa':'アフリカ','Asia':'アジア','Cruise Liner':'クルーズ船','Europe':'欧州','Latin America and Caribbean':'南米','Middle East':'中東',
     'North America':'北米','Oceania':'オセアニア'})
```

```python
countries[countries.Numeric.isna()]
```

```python
who = pd.read_csv('https://covid19.who.int/WHO-COVID-19-global-data.csv',na_filter=False)
who.columns = who.columns.str.replace(' ','')
```

```python
who_region = who[['Country_code','WHO_region']].drop_duplicates()
```

```python
countries['Region_who'] = countries['ISO2'].map(who_region.set_index('Country_code').WHO_region.to_dict())
```

```python
countries[countries.Region_who.isna()]
```

```python
countries['population'] = countries.Numeric.map(un.set_index('LocID').PopTotal.to_dict())
```

```python
countries = countries.append(pd.Series({'ISO2':'NaN','ISO3':'NaN','Country':'Cruise Liner','Region':'Cruise Liner',
                      'Classification':'Other*','Japanese':'クルーズ船などその他','latlong':'34.000,143.000','Region_j':'クルーズ船',
                     'Numeric':'NaN'}),ignore_index=True)
```

```python
countries
```

```python
#https://unstats.un.org/unsd/methodology/m49/overview/
unsd = pd.read_csv('UNSD — Methodology.csv')
```

```python
countries['ldc'] = countries['ISO3'].map(unsd.set_index('ISO-alpha3 Code')['Least Developed Countries (LDC)'].to_dict())
```

```python
countries
```

```python
eu = ['アイルランド','イタリア','エストニア','オーストリア','オランダ','キプロス','ギリシャ',
     'クロアチア','スウェーデン','スペイン','スロバキア','スロベニア','チェコ','デンマーク',
     'ドイツ','ハンガリー','フィンランド','フランス','ブルガリア','ベルギー','ポーランド',
     'ポルトガル','マルタ','ラトビア','リトアニア','ルーマニア','ルクセンブルク','英国']
```

```python
l = []

for i in countries.Japanese:
    if i in eu:
        l.append('EU')
    else:
        l.append(np.nan)
```

```python
countries['EU_dummy'] = l
```
